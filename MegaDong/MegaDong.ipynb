```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('Dong_Data_all.csv')
df.head()
```

```python
# df = df[['game_date','player_name','home_team','batter','pitcher','plate_z','hc_x','hc_y','hit_distance_sc','launch_angle']]
# df.head()
```

```python
df['spray_angle'] = np.tan((df['hc_x']-128)/(208-df['hc_y']))*180/np.pi*0.75

def spray_angle_cleanse(x):
    if x > 60:
        return 60
    if x < -60:
        return -60
    else:
        return x

df['spray_angle'] = df['spray_angle'].apply(lambda x: spray_angle_cleanse(x),1)
df.head()

```

```python
conditions = [
    (df['spray_angle']<=-46.666),
    (df['spray_angle']<=-33.333) & (df['spray_angle']>-46.666),
    (df['spray_angle']<=-20.000) & (df['spray_angle']>-33.333),
    (df['spray_angle']<=-6.666) & (df['spray_angle']>-20.000),
    (df['spray_angle']<=6.666) & (df['spray_angle']>-6.666),
    (df['spray_angle']<=20.000) & (df['spray_angle']>6.666),
    (df['spray_angle']<=33.333) & (df['spray_angle']>20.000),
    (df['spray_angle']<=46.666) & (df['spray_angle']>33.333),
    (df['spray_angle']>46.666)]

values = [1,2,3,4,5,6,7,8,9]
df['section'] = np.select(conditions,values)
df.head()
```

```python
df_most_common = df.groupby(['batter'])['section'].agg(
    lambda x: pd.Series.mode(x)[0])
df_most_common = df_most_common.to_frame().reset_index()
df_most_common = df_most_common.rename(columns={"section":"Most_common_section"})
```

```python
alt = pd.read_csv('Altitude_adj.csv')
alt.head()
```

```python
df2 = pd.merge(df,alt,left_on="home_team",right_on="Team")
df2.head()
```

```python
df2['adj_hit_distance'] = df2['hit_distance_sc']+(df2['hit_distance_sc']*df2['Adj Percent'])
df2.head()
```

```python
df2['V'] = np.sqrt(df2['adj_hit_distance']/((np.sin(2*df2['launch_angle']*np.pi/180))/32.174))
df2.head()
```

```python
df2['Vox'] = np.cos(df2['launch_angle']/180*np.pi)*df2['V']
df2.head()
```

```python
df2['Voy'] = np.sin(df2['launch_angle']/180*np.pi)* df2['V']
df2.head()
```

```python
dim = pd.read_csv('park_dim.csv')
dim.head()
```

```python
df_most_common.head()
```

```python
df2 = pd.merge(df2,df_most_common,left_on="batter",right_on="batter")
df2.head()
```

```python
df2['section'] = np.where(df2['section'] == 0,df2['Most_common_section'],df2['section'])
#df2 = df2.drop(['Most_common_section_x','Most_common_section_y'], axis = 1)
```

```python
df3 = pd.merge(df2,dim,left_on="section",right_on="Section")
df3.head()
```

```python
df3 = df3.drop(columns=["Section","Distance - Field","Height - Field"])
df3.head()
```

```python
df3['t'] = df3['Distance']/df3['Vox']
df3.head()
```

```python
df3['height_t'] = df3['plate_z']+(df3['Voy']*df3['t'])+0.5*(-32.174)*(df3['t']*df3['t'])
df3.head()
```

```python
df3.loc[df3['height_t'] <= df3['Height'], 'mega_dong'] = 0
df3.loc[df3['height_t'] > df3['Height'], 'mega_dong'] = 1
df3.head()
```
